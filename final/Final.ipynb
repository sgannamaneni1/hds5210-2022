# Factors Affecting Life Expectancy Rate

In this project, I have taken the life expectancy dataset to visualize and understand the data to get an idea/knowledge on the factors influencing the life expectancy rate. Below are the steps implemented as part of this project:

# Category: Data Access and Formats

```python
#importing pandas
#Reading Data File - 1(Format: csv)
#Source: Downloaded from kaggle.(Link: "https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who")
import pandas as pd
life_df= pd.read_csv("Life_Expectancy_Data.csv")
life_df.shape
```

```python
#Reading Data File - 2(Format: Excel)
#Source: Local file
# This file has suicide rate of male data 
male_df= pd.read_excel("Male.xlsx")
male_df.shape
```

```python
#Reading Data File - 3(Format: Excel)
#Source: Local file
# This file has suicide rate of Female data 
female_df= pd.read_excel("Female.xlsx")
female_df.shape
```

# Category: Data Merging

```python
#Merging data file - 1 & 2 and stored the result into new datdframe merge_1
merge_1 = pd.concat([life_df,male_df], axis=1, join='inner')
display(merge_1)
```

```python
#Merging data file - 1 & 3 and stored the result into new datdframe merge_2
merge_2 = pd.concat([life_df,female_df], axis=1, join='inner')
display(merge_2)
```

# Category: Data Transformation, Data Aggregation/Pivoting

```python
# Here the Life expectancy column data type of 'life_df' dataframe is transformed from object to float
# Created new data frame from 'life_df' with columns 'Life expectancy' & 'Country'
# Used pivot_table method to get mean, median, min, sum of life expectancy for each country

life_df['Life expectancy '] = life_df['Life expectancy '].astype(str).astype(float)

data = pd.DataFrame(life_df, columns= ['Life expectancy ','Country'])

pivot_1 = data.pivot_table(index=['Country'], values=['Life expectancy '], aggfunc={'sum','median','mean','min'})

print (pivot_1)
```

```python
# Removed duplicate columns from merge_1
# Here the Male_Suicide_Rate column data type of 'life_male' data frame is transformed from object to float
# Created new data frame from 'life_male' with columns 'Male_Suicide_Rate' & 'Country'
# Used pivot_table method to get mean, median, min, sum of Male_Suicide_Rate for each country

life_male=merge_1.T.drop_duplicates().T

life_male['Male_Suicide_Rate'] = life_male['Male_Suicide_Rate'].astype(str).astype(float)

data = pd.DataFrame(life_male, columns= ['Male_Suicide_Rate','Country'])

pivot_2 = data.pivot_table(index=['Country'], values=['Male_Suicide_Rate'], aggfunc={'sum','median','mean','min'})

print (pivot_2)
```

```python
# Removed duplicate columns from merge_2
# Here the Female_Suicide_Rate column data type of 'life_female' data frame is transformed from object to float
# Created new data frame from 'life_female' with columns 'Female_Suicide_Rate' & 'Country'
# Used pivot_table method to get mean, median, min, sum of Female_Suicide_Rate for each country

life_female=merge_2.T.drop_duplicates().T

life_female['Female_Suicide_Rate'] = life_female['Female_Suicide_Rate'].astype(str).astype(float)

data = pd.DataFrame(life_female, columns= ['Female_Suicide_Rate','Country'])

pivot_3 = data.pivot_table(index=['Country'], values=['Female_Suicide_Rate'], aggfunc={'sum','median','mean','min'})

print (pivot_3)
```

```python
# Dropped the rows with empty values from 'life_df' data frame to have accurate data
# Population column data type from float to int
# Created new data frame from 'life_df' with columns 'Population','infant deaths' & 'Country'
# Used pivot_table method to get mean, median, min, sum of population and infant deaths for each country

life_df.dropna(inplace = True)

life_df = life_df.astype({"Population":'int'})

data = pd.DataFrame(life_df, columns= ['Population','infant deaths','Country'])

pivot_4 = data.pivot_table(index=['Country'], values=['Population','infant deaths'], aggfunc={'sum','median','mean','min'})

print (pivot_4)
```

```python
# Created new data frame from 'life_df' with columns 'Life expectancy' & 'Status'
# Used pivot_table method to get mean, median, min, sum of life expectancy for developed and developing country

data = pd.DataFrame(life_df, columns= ['Life expectancy ','Status'])

pivot_5 = data.pivot_table(index=['Status'], values=['Life expectancy '], aggfunc={'sum','median','mean','min'})

print (pivot_5)
```

```python
#Imported all the necessary libraries to implement visual representation like Bar Chart, Pie Chart and Scatterplot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Below function is created to reuse the code to implement bar chart for various features
def barplot(df1,df2,XLabel,YLabel,title):
    plt.figure(figsize=(35,8))
    plt.title(title, fontsize=30)
    plt.ylabel(YLabel, fontsize=20)
    plt.xlabel(XLabel, fontsize=20)
    plt.xticks(rotation=90)
    plt.bar(df2,df1)
    plt.show()

#Below function is created to reuse the code to implement pie chart for various features   
def draw_pie_chart(df1, df2,new_df):
    df = new_df[[df1, df2]]
    df= df.groupby([df1]).agg({df2: sum})
    df.plot.pie(y=df2, figsize=(30, 30))

#Below function is created to reuse the code to implement scatter plot for various features
def Scatter_Plot(df1,df2):
    plt.figure(figsize=(15, 8))
    plt.scatter(df1, df2)
    plt.show()
```

```python
#Here df1 feature is the Y-axis and df2 feature is the X-axis
#XLabel and YLabel are the strings that need to be displayed in X and Y-axis. title is used to give a name to this visual representation.
#Finally, called the barplot function implemented in the previous cell
df1=pivot_1['Life expectancy ']['mean']
df2=pivot_1.index
XLabel="Countries"
YLabel="Life Expectancy"
title="Life Expectancy Rate by Each Country"
barplot(df1,df2,XLabel,YLabel,title)
```

Below I have created a class named Life_Expectancy and inside the class I have defined functions for life expectancy, infant deaths, male suicide rate and female suicide rate.

In Function life expectancy, I have given the below conditions-
1) if life expectancy rate is less than or equal to 50 then that country comes under poor category.
2) if life expectancy rate is greater than or equal to 50 and less than or equal to 60 then that country comes under moderate category.
3) if life expectancy rate is greater than or equal to 60 and less than or equal to 70 then that country comes under goods category.
4) if life expectancy rate is greater than or equal to 70 then that country comes under excellent category.

In Function infant deaths, I have given the below conditions-
1) if infant deaths is less than or equal to 30 then that country comes under poor category.
2) if infant deaths is greater than or equal to 30 and less than or equal to 50 then that country comes under moderate category.
3) if infant deaths is greater than or equal to 50 and less than or equal to 70 then that country comes under goods category.
4) if infant deaths is greater than or equal to 70 then that country comes under excellent category.

In Function male suicide rate, I have given the below conditions-
1) if male suicide rate is less than or equal to 3 then that country comes under poor category.
2) if male suicide rate is greater than or equal to 3 and less than or equal to 4 then that country comes under moderate category.
3) if male suicide rate is greater than or equal to 4 and less than or equal to 5 then that country comes under goods category.
4) if male suicide rate is greater than or equal to 5 then that country comes under excellent category.

In Function female suicide rate, I have given the below conditions-
1) if female suicide rate is less than or equal to 3 then that country comes under poor category.
2) if female suicide rate is greater than or equal to 3 and less than or equal to 4 then that country comes under moderate category.
3) if female suicide rate is greater than or equal to 4 and less than or equal to 5 then that country comes under goods category.
4) if female suicide rate is greater than or equal to 5 then that country comes under excellent category.


```python
class Life_Expectancy:
    def __init__(self,n):
        self.life = n
        
    def life_expec(self):
        if self.life <=50 :
            return "Poor"
        elif self.life >=50  and self.life <=60 :
            return "Moderate"
        elif self.life >=60 and self.life <=70:
            return "Good"
        elif self.life >=70:
            return "Excellent"
        
    def infant_death(self):
        if self.life <=30 :
            return "Excellent"
        elif self.life >=30  and self.life <=50 :
            return "Good"
        elif self.life >=50 and self.life <=70:
            return "Moderate"
        elif self.life >=70:
            return "Poor"
        
    def male_suicide_rate(self):
        if self.life <=3 :
            return "Excellent"
        elif self.life >=3  and self.life <=4 :
            return "Good"
        elif self.life >=4 and self.life <=5:
            return "Moderate"
        elif self.life >=5:
            return "Poor"
        
    def female_suicide_rate(self):
         if self.life <=3 :
            return "Excellent"
         elif self.life >=3  and self.life <=4:
            return "Good"
         elif self.life >=4 and self.life <=5:
            return "Moderate"
         elif self.life >=5:
            return "Poor"
```
